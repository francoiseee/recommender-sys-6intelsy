# EDA: MIND-Style News Recommendation Dataset

This notebook provides a quick exploratory analysis of `news.tsv` and `behaviors.tsv` used by the contextual bandit recommender.

```python
from pathlib import Path
import json
import pandas as pd

DATA_DIR = Path('../data')
NEWS_PATH = DATA_DIR / 'news.tsv'
BEHAVIORS_PATH = DATA_DIR / 'behaviors.tsv'

news_cols = ['news_id', 'category', 'subcategory', 'title', 'abstract', 'url', 'title_entities', 'abstract_entities']
beh_cols = ['impression_id', 'user_id', 'timestamp', 'history', 'impressions']
```

```python
news = pd.read_csv(NEWS_PATH, sep='\t', names=news_cols, usecols=range(8), engine='python')
beh = pd.read_csv(BEHAVIORS_PATH, sep='\t', names=beh_cols, usecols=range(5), engine='python')

print('News rows:', len(news))
print('Behavior rows:', len(beh))
news.head(3)
```

```python
news[['category', 'subcategory']].value_counts().head(10)
```

```python
beh['history_len'] = beh['history'].fillna('').map(lambda x: 0 if x == '' else len(str(x).split()))
beh['impression_len'] = beh['impressions'].fillna('').map(lambda x: 0 if x == '' else len(str(x).split()))

beh[['history_len', 'impression_len']].describe()
```
